```python
# Rest, split equally
import random
random.seed(113)
random.shuffle(assignmentsLeft)
random.shuffle(allVidsSet)

while (len(allVidsSet) > 0):
    whichAssignment = assignmentsLeft.pop()
    if whichAssignment == "Train":
        TrainSet.append(allVidsSet.pop())
    elif whichAssignment == "Valid":
        ValidSet.append(allVidsSet.pop())
    else:
        TestSet.append(allVidsSet.pop())

TrainSet.sort()
ValidSet.sort()
TestSet.sort()

TrainSetDuration = sum([vidLengths[thisVid] for thisVid in TrainSet])
ValidSetDuration = sum([vidLengths[thisVid] for thisVid in ValidSet])
TestSetDuration  = sum([vidLengths[thisVid] for thisVid in TestSet])
        
print "Train Length: " + str(TrainSetDuration) + ",  expected: " + str(int(0.60 * totalVidLengths)) + " diff: " + str(TrainSetDuration - int(0.60 * totalVidLengths))
print "Valid Length: " + str(ValidSetDuration) + ",    expected: " + str(int(0.20 * totalVidLengths)) + " diff: " + str(ValidSetDuration - int(0.20 * totalVidLengths))
print "Test  Length: " + str(TestSetDuration) + ",    expected: " + str(int(0.20 * totalVidLengths)) + " diff: " + str(TestSetDuration - int(0.20 * totalVidLengths))
```
